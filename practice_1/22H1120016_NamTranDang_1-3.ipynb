```python
# 4. Visualize mathplot histogram of 4 columns
import matplotlib.pyplot as plt # both matplot - seaborn plot use visualize graph
import seaborn as sns
import pandas as pd #panda -> manipulate data select,...

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'floral_species']
iris = pd.read_csv(url, sep=",",  names = column_names)


# set up the figure size - with 9 width and 5 height (inch) - the white canvase background ->if not use, this will be default of matplotlib
plt.figure(figsize=(9, 5))

# for loop [:-1] -> meaning from index 0 to 3 (not include 4)
# column_names [:-1] = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'] --> exclude 'floral_species'
# -> then enumerate (['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
# enumurate same as print normal array -> but enum can get index of array

for i, column in enumerate(column_names[0:1]):  # Remove collumn 'floral_species' -> don't need to plot since it not a number 
    
    # subplot -> 2 rows and 2 columns = 2x2 grids -> 4 subplots
    # devide 4 histogram in 4 corners of the white canvas 
    # matplot by default has index start from 1 
    # i+1 -> 1, 2, 3, 4 
    # 1 = top left, 2 = top right, 3 = bottom left, 4 = bottom right
    plt.subplot(2, 2, i+1)
    
    # plot histogram with 45 bins = 45 columns , kde = True (density curve on top) -> plot the density curve
    sns.histplot(iris[column], bins = 45, kde=True, color='blue')

    # tagging for histogram
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel("Frequency")

plt.tight_layout() #add space between subplots -> prevent overlapping
plt.show()  #show all plots at once 
```
